```python
from itertools import combinations

all_combinations = list(combinations(skill_clusters, 6))

# 중복이 되지 않는 경우를 찾기
desired_combinations = []

for combination in all_combinations:
    # 각 조합에서의 스킬을 모두 모아서 flatten
    flat_skills = [skill for sublist in combination for skill in sublist]

    # 중복이 되지 않는지 확인
    if all(flat_skills.count(skill) == 2 for skill in set(flat_skills)):
        desired_combinations.append(combination)

# 결과 출력
for idx, combination in enumerate(desired_combinations, start=1):
    print(f"{idx}. {combination}")
```

```python
def find_non_overlapping_combinations(combinations):
    non_overlapping_combinations = []

    for combo in combinations:
        first_words = [words[0] for words in combo]
        if len(set(first_words)) == len(first_words):
            non_overlapping_combinations.append(combo)

    return non_overlapping_combinations

result = find_non_overlapping_combinations(desired_combinations)
```
